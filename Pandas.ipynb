### Two Data Structures

```python
1. DataFrame:
    2D Data Structure >> Multiple Columns
    It consist of multiple 1D
    Table with rows and columns
    
2. Series:
    1D Data Structure >> Single Column
    Like and array
```

### Use Cases

```python
1. Read Data CSV, Excel, JSON
2. Read Data from Database(SQL and NoSQL)
3. Exploratory Data Analysis
4. Data Cleaning:
    Missing Values
    Outliers Handling
    
5. Encoding:
    1. Label Encoding
    2. One Hot Encoding
    
6. Data Preprocessing/Data Manipulation/Data Wrangling
7. Feature Engineering/Feature Importance/Feature Selection
8. Joining Multiple Tables (left, right, inner, outer)
9. Provides Time Series Functionality

Columns >> Features/Variables/ Attributes/ Parameteres/Columns
```
